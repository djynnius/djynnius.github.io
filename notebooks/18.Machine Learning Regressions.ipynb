```python
import numpy as np
import pandas as pd
from scipy.stats import linregress as lm 
from matplotlib import pyplot as plt 
import seaborn as sns

import statsmodels.formula.api as sm 
```

```python
school = pd.read_csv("school_scores.csv")
```

```python
school.head()
```

```python
correlation = school[['duration_of_study', 'math_score', 'further_maths_score']].corr()
```

```python
sns.heatmap(correlation)
```

```python
school[['sex', 'further_maths_score']].groupby('sex').mean()
```

```python
sns.boxplot(y='further_maths_score', x='sex', data=school)
```

```python
sns.boxplot(y='further_maths_score', x='class', data=school)
```

```python
sns.pairplot(school, hue='class')
```

```python
features = school[['duration_of_study', 'math_score']]
label = school['further_maths_score']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#help(train_test_split)
```

```python
features_training, features_test, label_training, label_test = train_test_split(features, label, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression #importing the algorithm
```

```python
algorithm = LinearRegression() #initializing the algorithm
```

```python
algorithm.fit(features_training, label_training) #training
```

```python
label_predicted = algorithm.predict(features_test)
```

```python
label_test[0:10]
```

```python
label_predicted[0:10]
```

```python
sns.scatterplot(label_test, label_predicted)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(label_test, label_predicted)
```

```python
np.sqrt(metrics.mean_squared_error(label_test, label_predicted))
```
