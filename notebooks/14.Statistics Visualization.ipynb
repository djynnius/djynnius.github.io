# Statistics Visualization

```python
import pandas as pd
from matplotlib import pyplot as plt #New module introduced
import seaborn as sns #New module introduced
```

```python
nhanes = pd.read_csv("nhanes.csv")
```

```python
#nhanes.info()
```

##### Histogram

```python
#nhanes['Age'].describe()
sns.histplot(x='Age', data=nhanes)
plt.show()
```

```python
#nhanes['BPDiaAve'].describe()
sns.histplot(x='BPDiaAve', data=nhanes)
plt.show()
```

##### Boxplots

```python
nhanes['Age'].describe()
```

```python
sns.boxplot(x='Age', data=nhanes)
plt.show()
```

### Categorical

##### Barplot (countplot)

```python
nhanes['Race1'].value_counts()
```

```python
sns.countplot(x='Race1', data=nhanes)
plt.show()
```

## Bivariate

##### Scatterplots (numeric vs numeric) - graphic representation of correlation

```python
sns.scatterplot(x='Age', y='Weight', data=nhanes)
plt.show()
```

```python
nhanes[['Age', 'Weight']].corr()
```

##### Grouped countplots

```python
sns.countplot(x='Gender', hue='Race1', data=nhanes)
plt.show()
```

##### Grouped Boxplots 

```python
sns.boxplot(y='Age', x='Race1', data=nhanes)
plt.show()
```

### Styling plots

```python
plt.figure(figsize=(10, 10)) #should come first
sns.boxplot(x='Age', y='Race1', data=nhanes)
plt.xlabel('Age (in years)')
plt.ylabel('Racial groups')
plt.title('Distribution of Age by Race')
plt.show()
```
